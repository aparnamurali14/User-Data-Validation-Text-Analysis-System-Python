```python
user_data = [" Anu, anu@gmail.com , 21 ","Ravi ,ravi123gmail.com, 17"," Suresh,suresh@gmail.com,25 ","Meena,meena@gmail,22",",invalid@gmail.com,30"]
newuser_data = []
for i in user_data:
    clean = i.replace(" ", "")
    split_data=clean.split(",")
    split_data[0]=split_data[0].title()
    split_data[1]=split_data[1].lower()
    newuser_data.append(split_data)
valid_users=[]
invalid_users=[]
underage_users=[]
eligible_users=[]
domain_count={}
domain_users=[]
for i in newuser_data:
    name=i[0]
    email=i[1]
    age=int(i[2])
    if age<18:
        underage_users.append(i)
    if email.count("@")==1 and email[0]!="@" and email[-1]!="@" and "." in email.split('@')[1]:
        domain_users.append(i)
    if email.count("@")==1 and email[0]!="@" and email[-1]!="@" and "." in email.split('@')[1] and age>0 and name!="":
        valid_users.append(i)
    else:
        invalid_users.append(i) 
    if age>18:
        eligible_users.append(i)     
for i in domain_users:
    email = i[1]
    domain = email.split("@")[1]
    if domain in domain_count:
        domain_count[domain] += 1  
    else:
        domain_count[domain] = 1 
 
print("\nValid_users(users with valid info)")
for i in valid_users:
    print(f"Name: {i[0]} | Email: {i[1]} | Age: {i[2]}") 
print("\nInvalid_users(users who don't have valid info)")
for i in invalid_users:
    print(f"Name: {i[0]} | Email: {i[1]} | Age: {i[2]}") 
print("\nUnderage_users(users whose age<18)")
for i in underage_users:
    print(f"Name: {i[0]} | Email: {i[1]} | Age: {i[2]}")
print("\nEligible_users(users whose age>=18)")
for i in eligible_users:
    print(f"Name: {i[0]} | Email: {i[1]} | Age: {i[2]}")
print("\nSummary Report")  
print("----------------")
print("Total number of records=",len(user_data)) 
print("Total valid_users=",len(valid_users))
print("Total invalid_users=",len(invalid_users))
print("Total underage_users(age<18)=",len(underage_users))
print("Total eligible_users(age>=18)=",len(eligible_users))
print("\nDomain summary")
print("-----------")
for i in domain_count:
    print("Domain:",i, "=", domain_count[i])
```
